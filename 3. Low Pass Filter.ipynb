```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

```python
pass_bands = int(input("Enter the pass band: "))
stop_bands = int(input("Enter the stop band: "))
fs = int(input("Enter the sampling frequency: "))
m = int(input("Enter the filter order: "))
```

```python
filter = sig.firwin(m,pass_bands,pass_zero='lowpass',window='hamming',fs=fs)
response = sig.freqz(filter,worN=8000)
plt.plot(fs*response[0]/2/np.pi,np.abs(response[1]),label='frequency response')
plt.grid()
plt.title('Low Pass filter')
plt.xlabel('frequency')
plt.ylabel('gain')
plt.show()
```
