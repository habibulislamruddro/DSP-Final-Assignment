```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
```

```python
pass_band = [.2,.35]
tran_width = .425-.1

filter_length = 67
```

```python
taps = signal.firwin(filter_length,pass_band,pass_zero='bandpass',window='blackman',width=tran_width)
response = signal.freqz(taps,worN=8000)
```

```python
taps = signal.firwin(67,pass_band,pass_zero='bandpass',window='blackman',width=tran_width)
response = signal.freqz(taps,worN=8000)
```

```python
plt.plot(response[0]/2/np.pi,np.abs(response[1]))
plt.grid()
```
