```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

```python
f1 = int(input("Enter the first pass band band: "))
f2 = int(input("Enter the second pass band: "))
stop_bands = int(input("Enter the stop band: "))
fs = int(input("Enter the sampling frequency: "))
m = int(input("Enter the filter order: "))
```

```python
#parameters
pass_bands = f1,f2
```

```python
filter = sig.firwin(m,pass_bands,pass_zero='bandstop',window='hamming',fs=fs)
response = sig.freqz(filter,worN=8000)
```

```python
plt.plot(fs*response[0]/2/np.pi,np.abs(response[1]),label='frequency response')
plt.grid()
plt.title('Band Stop Filter')
plt.xlabel('frequency')
plt.ylabel('gain')
plt.show()
```
