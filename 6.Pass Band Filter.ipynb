```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

```python
#parameters
pass_bands = 1500,2000
#stop_bands = 500
fs=10000
m=31
```

```python
filter = sig.firwin(m,pass_bands,pass_zero='bandpass',window='hamming',fs=fs)
response = sig.freqz(filter,worN=8000)
```

```python
plt.plot(fs*response[0]/2/np.pi,np.abs(response[1]),label='frequency response')
plt.grid()
plt.title('Band Pass Filter')
plt.xlabel('frequency')
plt.ylabel('gain')
plt.show()
```
