```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter,lfilter,freqz
```

```python
fs=100
time = np.arange(0,1,1/fs)
s= np.sin(2*np.pi*5*time)+np.sin(2*np.pi*15*time)+np.sin(2*np.pi*30*time)
```

```python
plt.figure(figsize=(10,4))
plt.plot(time,s)
plt.grid()
plt.show()
```

```python
def butter_stopband_filter(data,lowcut,highcut,fs,order=4):
    nyquist= 0.5*fs
    low = lowcut/nyquist
    high = highcut/nyquist
    b,a = butter(order,[low,high],btype='bandpass')
    y= lfilter(b,a,data )
    return y
```

```python
filtered_signal= butter_stopband_filter(s,10,25,fs )
```

```python
plt.figure(figsize=(10,4))
plt.plot(time,filtered_signal)
plt.title('Filtered Signal(Butterworth Bandstop Filter)')
plt.grid()
plt.show()
plt.show()
```
